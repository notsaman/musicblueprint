# Retrieves and prepares lyrics from genius.com for the LDA process

```python
import lyricsgenius
import pandas as pd
from tqdm.notebook import tqdm
```

```python
genius = lyricsgenius.Genius("DAyYDbP0dcrPUajZHsGN3hGmW0OjztM2QUTe0f5lEoycAwG32uJWs8frQLRciAPW")
genius.remove_section_headers = True
```

```python
df = pd.read_csv("song_features")
df = df.iloc[0:200]
```

```python
all_lyrics = []

for artist, track in tqdm(zip(df["Artist"], df["Track Name"])):
    song = genius.search_song(track, artist)
    if song is not None:
        lyrics = song.lyrics
        lyrics lyrics.replace('\n', ' ')
    else: 
        lyrics = ""
    entry = {"artist" : artist,
             "track" : track,
             "lyrics" : lyrics}
    all_lyrics.append(entry)
    
ly = pd.DataFrame(all_lyrics)
ly.to_csv("lyrics")
```
