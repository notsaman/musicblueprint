# Retrieves and prepares lyrics from genius.com for the LDA process

```python
import lyricsgenius
import pandas as pd
from tqdm.notebook import tqdm
```

```python
genius = lyricsgenius.Genius("DAyYDbP0dcrPUajZHsGN3hGmW0OjztM2QUTe0f5lEoycAwG32uJWs8frQLRciAPW")
genius.remove_section_headers = True
```

```python
#df = pd.read_csv("song_features") #Spotify top 200
#df = df.iloc[0:200]
df = pd.read_csv("rs100_songs") #Rolling Stones songs
sample = df.sample(200)
```

### Script for the RS artists

```python
import time
#df = df.iloc[:66]
df = sample
df = sample[46:]
all_lyrics2 = []

for artist, track in tqdm(zip(df["artist"], df["track"])):
    time.sleep(5)
    song = genius.search_song(track, artist)
    if song is not None:
        lyrics = song.lyrics
        lyrics = lyrics.replace('\n', ' ')
    else: 
        lyrics = ""
    entry = {"artist" : artist,
             "track" : track,
             "lyrics" : lyrics}
    all_lyrics.append(entry)
    
ly = pd.DataFrame(all_lyrics)
ly.to_csv("lyrics_rs100_v2")
```
