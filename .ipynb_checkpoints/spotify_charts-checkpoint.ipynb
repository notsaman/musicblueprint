# Retrieves list of songs from SpotifyCharts

```python
from fycharts.SpotifyCharts import SpotifyCharts
import pandas as pd
```

```python
api = SpotifyCharts()
api.top200Daily(output_file = "top_200_daily.csv", region = ["us"])
```

```python
df = pd.read_csv("top_200_daily.csv")
```
